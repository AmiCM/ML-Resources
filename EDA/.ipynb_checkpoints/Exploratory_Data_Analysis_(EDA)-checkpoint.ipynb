# Exploratory Data Analysis

En este notebook introduciremos algunas técnicas de exploración de datos, la idea es explorar un dataset con las películas mejor punteadas de netflix.

```python
import pandas as pd
import matplotlib.pyplot as plt #modulo visualizacion basica
import seaborn as sns #modulo visualizcion intermedia
```

```python
#cargar el csv
df = pd.read_csv('best_movies_netflix.csv')
```

```python
df
```

```python
#creando una variable categorica sobre la duracion de la pelicula
df['etiqueta_duracion'] = pd.cut(df['duration'],
                      bins=[0,100,150,float('inf')],
                      labels=['corta', 'media', 'larga'],
                      include_lowest =True)

df['etiqueta_score'] = pd.cut(df['score'],
                      bins=[6,7,8,float('inf')],
                      labels=['regular', 'buena', 'muy buena'],
                      include_lowest=True)
```

```python
df
```

```python
#agrupando y contando por etiqueta
df.groupby(['etiqueta_score'])['etiqueta_score'].count()
```

## Visualizaciones de variables en una dimensión

```python
#utiliznado pandas para visualizar un conjunto de histogramas
df.hist(bins=15, color='green', edgecolor='black', linewidth=1.0,
              xlabelsize=8, ylabelsize=8, grid=True)    
plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```

### Variables contínuas y categóricas

```python
fig = plt.figure(figsize = (6, 4))
title = fig.suptitle("Duración de las películas", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax1 = fig.add_subplot(1,1, 1)
ax1.set_xlabel("Duración")
ax1.set_ylabel("Density") 
#usando kdeplot de seaborn, esta funcion grafica la densidad estimada por kernel
sns.kdeplot(df['duration'], ax=ax1, shade=True, color='red')
```

```python
fig = plt.figure(figsize = (6, 4))
title = fig.suptitle("Duración (categórica)", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax = fig.add_subplot(1,1, 1)
ax.set_xlabel("Duración")
ax.set_ylabel("Frequencia") 
w_q = df['etiqueta_duracion'].value_counts()
w_q = (list(w_q.index), list(w_q.values))
ax.tick_params(axis='both', which='major', labelsize=8.5)
bar = ax.bar(w_q[0], w_q[1], color='steelblue', 
        edgecolor='black', linewidth=1)
```

### Correlation plot

```python
f, ax = plt.subplots(figsize=(10, 6))
#utilizando la funcion corr() de pandas
corr = df.corr()
#heatmap se utiliza para graficar la matriz de correlaciones
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
            linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Correlation Heatmap', fontsize=14)
```

```python
#utilizando seaborn para pairwise plots

cols = ['release_year','score','number_of_votes','duration']
pp = sns.pairplot(df, height=1.8, aspect=1.8, hue='etiqueta_score',
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Pairwise Plots', fontsize=14)
```

### ¿Estandarizar o no estandarizar?

```python
#tomando solamente variables numericas
subset_df = df[cols]

#cargando la funcion StandardScaler de la libreria preprocessing del modelo sklearn
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
#estandarizando los datos y asignandolos a scaled_df
scaled_df = ss.fit_transform(subset_df)
scaled_df = pd.DataFrame(scaled_df, columns=cols)
final_df = pd.concat([scaled_df, df['etiqueta_score']], axis=1)

#pairplot
pp2 = sns.pairplot(final_df, height=1.8, aspect=1.8, hue='etiqueta_score',
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp2.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Pairwise Plots', fontsize=14)
```

### Parallel coordinate plots (Inselberg 1985, Wegman 1990).
Usadas para examinar variables múltiples, correlaciones, agrupamientos en alta dimensión y otras relaciones. Se construye tomando los ejes paralelos en vez de la orientación ortogonal tradicional del sistema cartesiano. Cada observación está representada por una linea

```python
#usando pandas para parallel plots
from pandas.plotting import parallel_coordinates

plt.figure(figsize = (14,9))
pc = parallel_coordinates(frame = final_df, class_column = 'etiqueta_score')
```

### Scatterplots y distribuciones marginales

```python
#utilizando matplotlib.pyplot
plt.figure(figsize = (8,8))
plt.scatter(df['score'], df['release_year'],
            alpha=0.4, edgecolors='w')

plt.xlabel('Score')
plt.ylabel('Release year')
```

```python
#utiliznado seaborn para un joinplot
#CONCLUSION: LA DURACION Y EL SCORE NO PARECEN TENER UNA ALTA CORRELACION
#sns.set(rc={'figure.figsize':(8,8)})
jp = sns.jointplot(x='duration', y='score', data=df, kind='reg', space=0, height=6, ratio=4)
```

### Variables numéricas y categóricas. El Box Plot

```python
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))


sns.boxplot(x="duration", y="etiqueta_score", data=df,  ax=ax)
ax.set_xlabel("Duración",size = 12,alpha=0.8)
ax.set_ylabel("Etiqueta score",size = 12,alpha=0.8)
```

### Variables numéricas y categóricas. El Violin Plot
Parecido al boxplot, pero muestra la distribución estimada de la variable analizada

```python
f, (ax) = plt.subplots(1, 1, figsize=(12, 4))

sns.violinplot(x="etiqueta_score", y="duration", data=df,  ax=ax)
ax.set_xlabel("Etiqueta score",size = 12,alpha=0.8)
ax.set_ylabel("Duración",size = 12,alpha=0.8)
```

### también, podemos mostrar estimaciones de la densidad en 2 dimensiones e incluir una variable categórica a través del color

```python
#tomando solamente regular y y muy buena
df_regular = df[df['etiqueta_score'] == 'regular']
df_muy_buena = df[df['etiqueta_score'] == 'muy buena']
```

```python
#utilizando seaborn para plotear en azul la densidad de las regulares, y en rojo la densidad de las muy buenas
ax = sns.kdeplot(df_regular['score'], df_regular['duration'],cmap="Blues", shade=True, shade_lowest=False)
ax = sns.kdeplot(df_muy_buena['score'], df_muy_buena['duration'],cmap="Reds", shade=True, shade_lowest=False)
```
