# Introducción a Pandas

Pandas es un módulo de python desarrollado para trabajar con datos en formato data.frame

Su primera versión fue desarrollada por Wes McKinney en 2008 en formato open source.

El nombre "pandas" proviene del término "panel data", que es un término econométrico para designar a los datos que combinan una dimensión temporal con otra transversal. También es un acrónimo de la expresión "python data analysis".

## Documentación oficial

La documentación oficial de pandas se encuentra en la página:
https://pandas.pydata.org/docs/

Ahí se encuentran diversas guías y tutoriales para continuar aprendiendo.

Libros recomendados para aprender pandas:
   * "Python for Data Analysis: Data Wrangling with Pandas, NumPy, and IPython" de Wes McKinney.
   * "Python Data Science Handbook: Essential Tools for Working with Data" de Jake VanderPlas.
   * "Pandas for Everyone: Python Data Analysis" de Daniel Y. Chen.

## Filtrando con pandas 

```python
#tomando las observaciones para las que el petal_width es mayor a 2.3
filtro1 = df[df['petal_width'] > 2.3]
```

```python
filtro1
```

```python
#filtrando todas las flores de especie setosa
#OBSERVA que .isin es computacionalmente costoso (lento)
df[df['species'].isin(['setosa','virginica'])]
```

```python
#filtrando por especie con True or False
df['species']=='virginica' 
```

```python
#sumando las estadisticas por especie
df.groupby('species').sum()
```

```python
#calculando la media por especie
df3 = df.groupby('species').mean()
df3 
```

```python
#creando un csv a partir del df3 
#NO OLVIDES ESCRIBIR LA EXTENSION DEL ARCHIVO (.csv)
df3.to_csv('df3.csv')
```

## List comprehension

```python
df
```

```python
#creando una lista vacia
petalo_2 = []

#iterando por columna
for row in df['petal_length']:
    #multiplicando cada petal_length por 2
    petalo_2.append(row*2)

#creando una nueva columna
df['petalo_2'] = petalo_2

#imprimiendo el df
df
```

```python
#utilizando list comprehension
df['petalo_2'] = [row*2 for row in df['petal_length']]
```

```python
df
```

```python
#utilizando list comprehension con if-else
df['petalo_2'] = [row*2 if row < 2 else row *3 for row in df['petal_length'] ] 
```

```python
df
```

```python
# Ejercicio 9
df2 = california_housing.data
df2["MedInt"] = [int(i) for i in df2['MedInc']]
display(df2.head())
df2["MedRound"] = [round(i,2) for i in df2['MedInc']]
df2.head()
df2["IncPerAge"] = df2.apply(lambda x: x.MedInc/x.HouseAge, axis=1)
df2.head()
```

```python
# Ejercicio 10
df2["MedIncClass"] = ["High" if i>5.0 else "Low" for i in df2['MedInc']]
df2["MedIncClass2"] = df2.apply(lambda x: "High" if x.MedInc>5.0 else "Low", axis=1)
df2.head()
```
